```python
import random
import json
```

```python
import os

HOVER_ORIGINAL_PATH = "/home/namb/hoangpv4/kg_fact_checking/data/other_datasets/hover/hover_train_release_v1.1.json"
with open(HOVER_ORIGINAL_PATH, "r") as f:
    hover_data = json.load(f)
hops = list(set([hop["num_hops"] for hop in hover_data]))
hover_by_num_hops = {hop: [] for hop in hops}
for hop in hops:
    for item in hover_data:
        if item["num_hops"] == hop:
            hover_by_num_hops[hop].append(item)

for type_ in hops:
    random.seed(42)
    hover_by_num_hops[type_] = random.sample(hover_by_num_hops[type_], 1000)

for key, values in hover_by_num_hops.items():
    print(key, len(values))
    for i, sample in enumerate(values):
        values[i] = {
            "claim": sample["claim"],
            "id": sample["uid"],
            "Label": [True if sample["label"] == "SUPPORTED" else False],
        }
os.makedirs("/home/namb/hoangpv4/kg_fact_checking/data/other_datasets/processed_hover", exist_ok=True)
with open("/home/namb/hoangpv4/kg_fact_checking/data/other_datasets/processed_hover/process_hover_train.json", "w") as f:
    json.dump(hover_by_num_hops, f, indent=4)
```

```python
FEVEROUS_ORIGINAL_PATH = "/home/namb/hoangpv4/kg_fact_checking/data/other_datasets/feverous/feverous_train_challenges.jsonl"
with open(FEVEROUS_ORIGINAL_PATH, "r") as f:
    feverous_data = [json.loads(line) for line in f.readlines()][1:]
    feverous_data = [item for item in feverous_data if item["label"] in ["REFUTES", "SUPPORTS"]]

types = ["Entity Disambiguation", "Multi-hop Reasoning"]
feverous_by_type = {type_: [] for type_ in types}

for type_ in types:
    for item in feverous_data:
        if item["challenge"] == type_:
            feverous_by_type[type_].append(item)

for type_ in types:
    random.seed(42)
    feverous_by_type[type_] = random.sample(feverous_by_type[type_], min(1000, len(feverous_by_type[type_])))

for key, values in feverous_by_type.items():
    print(key, len(values))
    for i, sample in enumerate(values):
        values[i] = {
            "claim": sample["claim"],
            "id": sample["id"],
            "Label": [True if sample["label"] == "SUPPORTS" else False],
        }
os.makedirs("/home/namb/hoangpv4/kg_fact_checking/data/other_datasets/processed_feverous", exist_ok=True)
with open("/home/namb/hoangpv4/kg_fact_checking/data/other_datasets/processed_feverous/process_feverous_train.json", "w") as f:
    json.dump(feverous_by_type, f, indent=4)
```
