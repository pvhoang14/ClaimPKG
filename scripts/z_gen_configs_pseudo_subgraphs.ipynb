### Check performance

```python
path = "/home/namb/hoangpv4/kg_fact_checking/data/output/pseudo_subgraphs"
from glob import glob
import os
import json
import random

for file in sorted(glob(os.path.join(path, "*", "*.json"))):
    finished = True
    print("=* " * 10 + "||" + file.split("/")[-1].upper() + "||" + "*= " * 10)
    print()
    with open(file, "r") as f:
        data = json.load(f)["data"]
        for i, (k, v) in enumerate(data.items()):
            sample = random.choice(v)
            if "pseudo_subgraphs" not in sample:
                continue

            print(f"{i}) " + k.upper())
            print(f"Claim:", sample['claim'])
            print(f"Entities:", sample["Entity_set"])
            for j, graph in enumerate(sample["pseudo_subgraphs"]):
                print(f"Graph: {j}")
                for triplet in graph.split("\n"):
                    print(f"    {triplet}")
            print()
    print("\n")          
```
